# Hill Equation

Enzymatic reactions are usually mathematically modeled using a fuction called the "Hill equation", which describes the rate of production of some product Y, as a function of the concentration of the catalyst X. Depending on its parameters, the Hill equation can vary from a simple hyperbolic curve to a sigmoidal function.

The Hill equation for an activator is of the form:

$$
f(X) = \frac{\beta X^n}{K^n + X^n}
$$

where $X$ represents the concentration of a transcriptional activator and $f(X)$  represents the the combined transcription and translation of the gene $Y$ that is regulated by $X$.

If rather than stimulating the production of $Y$, $X$ "represses" $Y$, we can write the corresponding Hill function as:

$$
f(X) = \frac{\beta}{1 + (X/K)^n}
$$




# An interactive visualization of the Hill Equation

```python
# import statements to make numeric and plotting functions available
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
from IPython.html import widgets


def hill_activating(X, B, K, n):
    """ Hill function for an activator"""
    return (B * X**n)/(K**n + X**n)

def hill_repressing(X, B, K, n):
    """ Hill function for a repressor"""
    return B/(1.0 + (X/K)**n)


# generate a range of x values, representing a range of concentrations of our
# transcription factor X
x = linspace(0,35,200)  # generate 250 evenly spaced points between 0 and 35

def hill_activating_plot(n,B,K):
    # calculating corresponding rates of production of Y
    y_activation = hill_activating(x, B, K, n) 
    y_repression = hill_repressing(x, B, K, n)

    # plot the hill fxn with the user set parameters
    fig, (ax1,ax2) = subplots(1, 2, figsize=(10,4), sharex=True, sharey=True)
    
    ax1.plot(x, y_activation, label='B = {}, K = {}, n={}'.format(B, K, n))
    ax1.set_ylim(-0.5,8)
    ax1.set_xlabel('Concentration of X')
    ax1.set_ylabel('Rate of production of Y')
    ax1.set_title('Hill function, activation')
    
    ax2.plot(x, y_repression, label='B = {}, K = {}, n={}'.format(B, K, n))
    ax2.set_xlabel('Concentration of X')
    ax2.set_title('Hill function, repression')    


widgets.interact(hill_activating_plot, 
                 n=widgets.IntSlider(min=1,max=10,step=1,value=1),
                 B=widgets.FloatSlider(min=1,max=7,step=0.1,value=5),
                 K=widgets.FloatSlider(min=1,max=25,step=0.1,value=10),)
pass
```
